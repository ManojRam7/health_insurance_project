```python
# ============================================================================
# STEP 2: Setup for Synthetic Data Generation
# ============================================================================

import random
from datetime import datetime, timedelta
from pyspark.sql.types import *

# Set seed for reproducibility
random.seed(42)

# Configuration
BATCH_SIZE_PROVIDERS = 100
BATCH_SIZE_POLICIES = 100
BATCH_SIZE_MEMBERS = 100
BATCH_SIZE_CLAIMS = 100

# Reference data that matches old data
REGIONS = ["London", "Manchester", "Birmingham", "Leeds", "Liverpool", "Sheffield", "Bristol", 
           "Edinburgh", "Cardiff", "Belfast", "Durham", "Leicester"]
PRODUCT_LINES = ["Health Basic", "Health Premium", "Preventive Care", "Family Coverage", "Corporate Health"]
CHANNELS = ["Direct", "Agent", "B2B", "Partner", "Online"]
CLAIM_TYPES = ["Outpatient", "Inpatient", "Dental", "Imaging", "Lab", "Pharmacy", "Hospital", "Maternity"]
CLAIM_STATUSES = ["Settled", "Pending", "Rejected", "Approved"]
EMPLOYMENT_STATUSES = ["Employed", "Self-Employed", "Unemployed", "Retired", "Student"]

print("✅ Configuration loaded")
print(f"   Batch: {BATCH_SIZE_PROVIDERS} providers, {BATCH_SIZE_POLICIES} policies, {BATCH_SIZE_MEMBERS} members, {BATCH_SIZE_CLAIMS} claims")
```

```python
# ============================================================================
# STEP 3: Load Max IDs to continue sequencing
# ============================================================================

# Get max IDs from existing data
max_provider_id = providers_df.select(F.max("Provider_ID")).collect()[0][0]
max_policy_id = policys_df.select(F.max("Policy_ID")).collect()[0][0]
max_member_id = members_df.select(F.max("Member_ID")).collect()[0][0]
max_claim_id = claims_df.select(F.max("Claim_ID")).collect()[0][0]

print(f"✅ Max IDs loaded:")
print(f"   Providers: {max_provider_id}")
print(f"   Policies: {max_policy_id}")
print(f"   Members: {max_member_id}")
print(f"   Claims: {max_claim_id}")

# Extract numeric portions for sequencing
import re
provider_num = int(re.search(r'\d+', max_provider_id).group()) if max_provider_id else 50000
policy_num = int(re.search(r'\d+', max_policy_id).group()) if max_policy_id else 0
member_num = int(re.search(r'\d+', max_member_id).group()) if max_member_id else 0
claim_num = int(re.search(r'\d+', max_claim_id).group()) if max_claim_id else 0

next_provider_num = provider_num + 1
next_policy_num = policy_num + 1
next_member_num = member_num + 1
next_claim_num = claim_num + 1

print(f"\n✅ Next IDs for batch:")
print(f"   Providers: PRV{next_provider_num:05d}")
print(f"   Policies: P_{next_policy_num:08d}")
print(f"   Members: MEM_{next_member_num:08d}")
print(f"   Claims: CLM{next_claim_num:06d}")
```

```python
# ============================================================================
# STEP 4: Generate Synthetic Providers (APPEND MODE)
# ============================================================================

print("\n🔧 Generating synthetic providers...")

# Strategy: APPEND new providers with new IDs (not referencing existing ones)
providers_data = []

for i in range(BATCH_SIZE_PROVIDERS):
    provider_id = f"PRV{next_provider_num + i:05d}"
    fraud_flag = 1 if random.random() < 0.0935 else 0  # 9.35% fraud rate
    
    providers_data.append({
        "Provider_ID": provider_id,
        "Fraud_Flag": fraud_flag
    })

providers_synthetic_df = spark.createDataFrame(providers_data, schema=providers_schema)

print(f"✅ Generated {len(providers_data)} synthetic providers")
print(f"   Range: {providers_data[0]['Provider_ID']} → {providers_data[-1]['Provider_ID']}")
print(f"   Fraud rate: {(sum(1 for p in providers_data if p['Fraud_Flag'] == 1) / len(providers_data) * 100):.2f}%")
```

```python
# ============================================================================
# STEP 5: Generate Synthetic Policies (USE EXISTING PROVIDERS)
# ============================================================================

print("\n🔧 Generating synthetic policies...")

# Get all existing provider IDs to maintain referential integrity
existing_providers = providers_df.select("Provider_ID").rdd.flatMap(lambda x: x).collect()
print(f"   📌 Using {len(existing_providers)} existing provider IDs for referential integrity")

policies_data = []

for i in range(BATCH_SIZE_POLICIES):
    policy_id = f"P_{next_policy_num + i:08d}"
    customer_id = f"CUST{random.randint(100000, 999999)}"
    product_line = random.choice(PRODUCT_LINES)
    sum_insured = float(round(random.uniform(100000, 5000000), 2))
    annual_premium = float(round(sum_insured * random.uniform(0.005, 0.06), 2))
    policy_start = datetime.strptime(
        f"{random.randint(2015, 2022)}-{random.randint(1, 12):02d}-{random.randint(1, 28):02d}",
        "%Y-%m-%d"
    ).date()
    policy_end = (datetime.combine(policy_start, datetime.min.time()) + timedelta(days=365)).date()
    renewal_offered = 1 if random.random() < 0.7491 else 0  # 74.91% renewal rate
    renewal_accepted = 1 if renewal_offered == 1 and random.random() < 0.75 else 0
    discount_pct = float(round(random.uniform(0, 25), 2))
    channel = random.choice(CHANNELS)
    
    policies_data.append({
        "Policy_ID": policy_id,
        "Customer_ID": customer_id,
        "Product_Line": product_line,
        "Sum_Insured_GBP": sum_insured,
        "Annual_Premium_GBP": annual_premium,
        "Policy_Start_Date": policy_start,
        "Policy_End_Date": policy_end,
        "Renewal_Offered_Flag": renewal_offered,
        "Renewal_Accepted_Flag": renewal_accepted,
        "Discount_Offered_Pct": discount_pct,
        "Channel": channel
    })

policies_synthetic_df = spark.createDataFrame(policies_data, schema=policies_schema)

print(f"✅ Generated {len(policies_data)} synthetic policies")
print(f"   Range: {policies_data[0]['Policy_ID']} → {policies_data[-1]['Policy_ID']}")
print(f"   Renewal offered rate: {(sum(1 for p in policies_data if p['Renewal_Offered_Flag'] == 1) / len(policies_data) * 100):.2f}%")
```

```python
# ============================================================================
# STEP 6: Generate Synthetic Members (USE EXISTING POLICIES)
# ============================================================================

print("\n🔧 Generating synthetic members...")

# Get all existing policy IDs (both old + new)
all_policy_ids = policys_df.select("Policy_ID").rdd.flatMap(lambda x: x).collect() + \
                  policies_synthetic_df.select("Policy_ID").rdd.flatMap(lambda x: x).collect()
print(f"   📌 Using {len(all_policy_ids)} existing policy IDs (old + new) for referential integrity")

members_data = []

for i in range(BATCH_SIZE_MEMBERS):
    member_id = f"MEM_{next_member_num + i:08d}"
    # Use EXISTING policy IDs to maintain referential integrity
    policy_id = random.choice(all_policy_ids)
    age = max(18, min(85, int(random.gauss(42, 15))))  # Gaussian distribution: mean=42, σ=15
    gender = random.choice(["M", "F"])
    bmi = round(random.uniform(18.5, 40.0), 2)
    smoker = random.choice(["Yes", "No"])
    chronic_disease = random.choice(["Diabetes", "Hypertension", "Asthma", "None"])
    employment_status = random.choice(EMPLOYMENT_STATUSES)
    region = random.choice(REGIONS)
    
    members_data.append({
        "Member_ID": member_id,
        "Policy_ID": policy_id,
        "Age": age,
        "Gender": gender,
        "BMI": bmi,
        "Smoker": smoker,
        "Chronic_Disease": chronic_disease,
        "Employment_Status": employment_status,
        "Region": region
    })

members_synthetic_df = spark.createDataFrame(members_data, schema=members_schema)

print(f"✅ Generated {len(members_data)} synthetic members")
print(f"   Range: {members_data[0]['Member_ID']} → {members_data[-1]['Member_ID']}")
print(f"   ✓ All members link to existing policies (100% referential integrity)")
print(f"   Avg age: {sum(m['Age'] for m in members_data) / len(members_data):.2f}")
print(f"   Avg BMI: {sum(m['BMI'] for m in members_data) / len(members_data):.2f}")
```

```python
# ============================================================================
# STEP 7: Generate Synthetic Claims (USE EXISTING PROVIDERS & MEMBERS)
# ============================================================================

print("\n🔧 Generating synthetic claims...")

# Get all existing provider IDs (both old + new)
all_provider_ids = providers_df.select("Provider_ID").rdd.flatMap(lambda x: x).collect() + \
                   providers_synthetic_df.select("Provider_ID").rdd.flatMap(lambda x: x).collect()

# Get all existing member IDs (both old + new)
all_member_ids = members_df.select("Member_ID").rdd.flatMap(lambda x: x).collect() + \
                 members_synthetic_df.select("Member_ID").rdd.flatMap(lambda x: x).collect()

print(f"   📌 Using {len(all_provider_ids)} existing provider IDs (old + new)")
print(f"   📌 Using {len(all_member_ids)} existing member IDs (old + new)")

claims_data = []

for i in range(BATCH_SIZE_CLAIMS):
    claim_id = f"CLM{next_claim_num + i:06d}"
    # Use EXISTING provider and member IDs
    provider_id = random.choice(all_provider_ids)
    member_key = random.choice(all_member_ids)
    date_reported = datetime.now().date() - timedelta(days=random.randint(0, 730))
    date_settled = date_reported + timedelta(days=random.randint(1, 60))
    claim_amount = float(round(random.uniform(50, 10000), 2))
    payout = float(round(claim_amount * random.uniform(0.6, 1.0), 2))
    fraud_label = 1 if random.random() < 0.0153 else 0  # 1.53% fraud rate
    claim_type = random.choice(CLAIM_TYPES)
    claim_status = random.choice(CLAIM_STATUSES)
    
    claims_data.append({
        "Claim_ID": claim_id,
        "Provider_ID": provider_id,
        "Member_Key": member_key,
        "Date_Reported": date_reported,
        "Date_Settled": date_settled,
        "Payout_GBP": payout,
        "Claim_Amount_GBP": claim_amount,
        "Fraud_Label": fraud_label,
        "Claim_Type": claim_type,
        "Claim_Status": claim_status
    })

claims_synthetic_df = spark.createDataFrame(claims_data, schema=claims_schema)

print(f"✅ Generated {len(claims_data)} synthetic claims")
print(f"   Range: {claims_data[0]['Claim_ID']} → {claims_data[-1]['Claim_ID']}")
print(f"   ✓ Fraud rate: ~{(sum(1 for c in claims_data if c['Fraud_Label'] == 1) / len(claims_data) * 100):.2f}%")
print(f"   ✓ All claims link to existing providers and members (100% referential integrity)")
```

```python
# ============================================================================
# STEP 8: PREVIEW - Sample data before appending
# ============================================================================

print("\n" + "=" * 100)
print("📊 PREVIEW OF SYNTHETIC DATA")
print("=" * 100)

print("\n🔷 PROVIDERS (first 10):")
providers_synthetic_df.limit(10).show(truncate=False)

print("\n🔶 POLICIES (first 10):")
policies_synthetic_df.select("Policy_ID", "Customer_ID", "Product_Line", "Sum_Insured_GBP", "Channel").limit(10).show(truncate=False)

print("\n🔷 MEMBERS (first 10):")
members_synthetic_df.select("Member_ID", "Policy_ID", "Age", "Gender", "Region").limit(10).show(truncate=False)

print("\n🔶 CLAIMS (first 10):")
claims_synthetic_df.select("Claim_ID", "Provider_ID", "Member_Key", "Payout_GBP", "Claim_Status").limit(10).show(truncate=False)

print("\n" + "=" * 100)
print("📈 QUALITY METRICS")
print("=" * 100)

total_batch = len(providers_data) + len(policies_data) + len(members_data) + len(claims_data)
fraud_prov = (providers_synthetic_df.filter("Fraud_Flag = 1").count() / len(providers_data)) * 100
fraud_claims = (claims_synthetic_df.filter("Fraud_Label = 1").count() / len(claims_data)) * 100

print(f"\n✅ Total records to append: {total_batch:,}")
print(f"   - Providers: {len(providers_data):,} (fraud: {fraud_prov:.2f}%)")
print(f"   - Policies: {len(policies_data):,}")
print(f"   - Members: {len(members_data):,}")
print(f"   - Claims: {len(claims_data):,} (fraud: {fraud_claims:.2f}%)")
print(f"\n✓ All referential integrity maintained")
print(f"✓ Ready to append to ADLS")
```

```python
# ============================================================================
# STEP 9: APPEND to ADLS Tables
# ============================================================================

print("\n" + "=" * 100)
print("🔄 APPENDING SYNTHETIC DATA TO ADLS")
print("=" * 100)

try:
    print("\n📤 Appending PROVIDERS...")
    providers_synthetic_df.write.format("delta").mode("append").option("mergeSchema", "true").saveAsTable(f"{DB_BRONZE}.providers")
    print(f"✅ Providers appended: {len(providers_data):,} records")
    
    print("\n📤 Appending POLICIES...")
    policies_synthetic_df.write.format("delta").mode("append").option("mergeSchema", "true").saveAsTable(f"{DB_BRONZE}.policies")
    print(f"✅ Policies appended: {len(policies_data):,} records")
    
    print("\n📤 Appending MEMBERS...")
    members_synthetic_df.write.format("delta").mode("append").option("mergeSchema", "true").saveAsTable(f"{DB_BRONZE}.members")
    print(f"✅ Members appended: {len(members_data):,} records")
    
    print("\n📤 Appending CLAIMS...")
    claims_synthetic_df.write.format("delta").mode("append").option("mergeSchema", "true").saveAsTable(f"{DB_BRONZE}.claims")
    print(f"✅ Claims appended: {len(claims_data):,} records")
    
    total_appended = len(providers_data) + len(policies_data) + len(members_data) + len(claims_data)
    print(f"\n" + "=" * 100)
    print(f"🎉 SUCCESS: {total_appended:,} records appended to ADLS!")
    print(f"=" * 100)
    
except Exception as e:
    print(f"\n❌ ERROR: {str(e)}")
    raise
```
