```python
from pyspark.sql import SparkSession

packages = ",".join([
    "io.delta:delta-spark_2.12:3.1.0",                 # Delta Lake
    "org.apache.hadoop:hadoop-azure:3.3.4",            # ADLS Gen2 connector
    "com.azure:azure-storage-blob:12.19.0",            # Blob SDK
    "com.azure:azure-identity:1.10.0"                  # Azure OAuth support
])

spark = (
    SparkSession.builder
        .appName("bupa-local-adls-delta")
        .master("local[*]")
        .config("spark.jars.packages", packages)
        .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
        .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
        .getOrCreate()
)

spark
```
