```python
# Create all gold layer tables
DB_BRONZE   = "bupa_bronze"

spark.sql(f"CREATE DATABASE IF NOT EXISTS {DB_BRONZE}")



tables = [
    "policies",
    "members",
    "claims",
    "providers"
]

for table in tables:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS {DB_BRONZE}.{table}
        USING DELTA
        LOCATION 'abfss://rawdata@clientdatastorage.dfs.core.windows.net/{table}'
    """)

print(f"Created {len(tables)} tables in '{DB_BRONZE}' database")
```
