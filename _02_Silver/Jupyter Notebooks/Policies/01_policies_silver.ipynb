```python
# Cell 2 — Read Bronze policies and enforce schema

# 1) Read bronze Delta
pol_bz = spark.read.format("delta").load(BRONZE_POLICIES_PATH)

print("Bronze raw schema:")
pol_bz.printSchema()

# 2) Select & cast to clean schema columns
pol = (
    pol_bz.select(
        F.col("Policy_ID").cast("string"),
        F.col("Customer_ID").cast("string"),
        F.col("Product_Line").cast("string"),
        F.col("Sum_Insured_GBP").cast("double"),
        F.col("Annual_Premium_GBP").cast("double"),
        F.to_date("Policy_Start_Date").alias("Policy_Start_Date"),
        F.to_date("Policy_End_Date").alias("Policy_End_Date"),
        F.col("Renewal_Offered_Flag").cast("int"),
        F.col("Renewal_Accepted_Flag").cast("int"),
        F.col("Discount_Offered_%").cast("double"),
        F.col("Channel").cast("string"),
    )
)

print("Bronze (selected) rowcount:", pol.count())

# 3) Enforce strict target schema
policy_schema = StructType([
    StructField("Policy_ID",             StringType()),
    StructField("Customer_ID",           StringType()),
    StructField("Product_Line",          StringType()),
    StructField("Sum_Insured_GBP",       DoubleType()),
    StructField("Annual_Premium_GBP",    DoubleType()),
    StructField("Policy_Start_Date",     DateType()),
    StructField("Policy_End_Date",       DateType()),
    StructField("Renewal_Offered_Flag",  IntegerType()),
    StructField("Renewal_Accepted_Flag", IntegerType()),
    StructField("Discount_Offered_%",    DoubleType()),
    StructField("Channel",               StringType()),
])

pol = enforce_schema(pol, policy_schema)

print("After enforce_schema:")
pol.printSchema()
```

# 🔹 Cell 3 — Key checks & quarantine

```python
# Cell 3 — Key checks (Policy_ID, Customer_ID) & quarantine

# Split OK vs BAD records by key presence
pol_bad = pol.filter(
    F.col("Policy_ID").isNull() | F.col("Customer_ID").isNull()
)

pol_ok = pol.filter(
    F.col("Policy_ID").isNotNull() & F.col("Customer_ID").isNotNull()
)

print("Total rows:", pol.count())
print("Rows with missing keys:", pol_bad.count())
print("Rows with valid keys:", pol_ok.count())

# Send bad rows to central quarantine table
if pol_bad.count() > 0:
    quarantine(pol_bad, "null_key_policy_or_customer", "policies")

print("✅ Key check & quarantine completed.")
```

# 🔹 Cell 4 — Business-rule DQ flags (money, discount, renewal)

```python
# Cell 4 — DQ flags: money, discount, renewal logic

pol_ok = (
    pol_ok
    .withColumn(
        "dq_money_valid",
        F.when(
            (F.col("Annual_Premium_GBP") >= 0) &
            (F.col("Sum_Insured_GBP") >= 0),
            1
        ).otherwise(0)
    )
    .withColumn(
        "dq_discount_valid",
        F.when(
            (F.col("Discount_Offered_%").between(0, 100)) |
            (F.col("Discount_Offered_%").isNull()),
            1
        ).otherwise(0)
    )
    .withColumn(
        "dq_renewal_valid",
        F.when(
            ~(
                (F.col("Renewal_Accepted_Flag") == 1) &
                (F.coalesce(F.col("Renewal_Offered_Flag"), F.lit(0)) != 1)
            ),
            1
        ).otherwise(0)
    )
)

print("Sample with DQ flags:")
pol_ok.select(
    "Policy_ID",
    "Annual_Premium_GBP",
    "Sum_Insured_GBP",
    "Discount_Offered_%",
    "Renewal_Offered_Flag",
    "Renewal_Accepted_Flag",
    "dq_money_valid",
    "dq_discount_valid",
    "dq_renewal_valid",
).show(10, truncate=False)
```

# 🔹 Cell 5 — Deduplicate, normalize, derive features

```python
# Cell 5 — Deduplicate + normalize categories + derived columns

# Deduplicate per Policy_ID (latest by Policy_Start_Date)
pol_silver = drop_dupes_keep_latest(
    pol_ok,
    key_cols=["Policy_ID"],
    order_desc_cols=["Policy_Start_Date"]
)

# Normalize categories (case & whitespace)
pol_silver = (
    pol_silver
    .withColumn("Product_Line", F.initcap(F.trim(F.col("Product_Line"))))
    .withColumn("Channel",      F.initcap(F.trim(F.col("Channel"))))
)

# Derived feature: Policy duration
pol_silver = pol_silver.withColumn(
    "Policy_Duration_Days",
    F.when(
        F.col("Policy_Start_Date").isNotNull() &
        F.col("Policy_End_Date").isNotNull(),
        F.datediff("Policy_End_Date", "Policy_Start_Date")
    ).otherwise(F.lit(None).cast("int"))
)

# Derived feature: Renewal_Conversion (only meaningful when offer=1)
pol_silver = pol_silver.withColumn(
    "Renewal_Conversion",
    F.when(F.col("Renewal_Offered_Flag") == 1,
           F.col("Renewal_Accepted_Flag"))
     .otherwise(F.lit(None).cast("int"))
)

print("Sample Silver policies (post-feature-engineering):")
pol_silver.show(10, truncate=False)
```

# 🔹 Cell 7 — Strong DQ expectations & dictionary checks

```python
# ==========================================
# Cell 7 — Data Quality (DQ) validations
# Uses dq_expect / dq_left_anti_ref from utils_silver
# ==========================================

from pyspark.sql import functions as F

# --- Normalize renewal flags to clean 0/1 ints ---
def to_flag(colname):
    s = F.lower(F.trim(F.col(colname).cast("string")))
    return (
        F.when(s.isin("1", "y", "yes", "true", "t", "on"), 1)
         .when(s.isin("0", "n", "no", "false", "f", "off"), 0)
         .otherwise(F.col(colname).cast("int"))
    )

pol_silver = (
    pol_silver
    .withColumn("Renewal_Offered_Flag",  to_flag("Renewal_Offered_Flag").cast("int"))
    .withColumn("Renewal_Accepted_Flag", to_flag("Renewal_Accepted_Flag").cast("int"))
)

# 1) Basic expectations (hard-error) — NOTE: matches utils_silver.dq_expect signature

dq_expect(
    pol_silver,
    "key_not_null",
    "Policy_ID IS NOT NULL AND Customer_ID IS NOT NULL",
    "error",
    "policies",
    paths_silver
)

dq_expect(
    pol_silver,
    "date_order",
    "Policy_Start_Date IS NULL OR Policy_End_Date IS NULL OR Policy_End_Date >= Policy_Start_Date",
    "error",
    "policies",
    paths_silver
)

dq_expect(
    pol_silver,
    "money_non_negative",
    "coalesce(Annual_Premium_GBP,0) >= 0 AND coalesce(Sum_Insured_GBP,0) >= 0",
    "error",
    "policies",
    paths_silver
)

dq_expect(
    pol_silver,
    "discount_bounds",
    "`Discount_Offered_%` IS NULL OR (`Discount_Offered_%` BETWEEN 0 AND 100)",
    "error",
    "policies",
    paths_silver
)

# 2) Renewal consistency — SLA-gated block (manual, not via dq_expect)

viol_df = pol_silver.filter(
    (F.col("Renewal_Accepted_Flag") == 1) &
    (F.coalesce(F.col("Renewal_Offered_Flag"), F.lit(0)) != 1)
)
viol_cnt = viol_df.count()
total    = pol_silver.count()
viol_pct = 0 if total == 0 else round(viol_cnt / total * 100.0, 4)

if viol_cnt > 0:
    # uses quarantine from utils_silver, which requires paths_silver
    quarantine(viol_df, "renewal_inconsistent", "policies", paths_silver)

print(f"[RECON] renewal_inconsistent: {viol_cnt}/{total} ({viol_pct}%)")

# Relaxed SLA to unblock; you can tighten later
SLA_PCT = 25.0
if viol_pct > SLA_PCT:
    raise Exception(f"DQ gate failed: policies · renewal_consistency {viol_pct}% > {SLA_PCT}%")
else:
    print("✅ DQ PASS [policies] renewal_consistency (within SLA)")

# 3) Dictionary-based validations (ERROR severity)
#    Uses dq_left_anti_ref from utils_silver, which also takes paths_silver

# dim_channel and dim_product_line should be defined in Cell 6
dq_left_anti_ref(
    pol_silver,
    dim_channel,
    "Channel",
    "Channel",
    "channel_in_dictionary",
    "error",
    "policies",
    paths_silver
)

dq_left_anti_ref(
    pol_silver,
    dim_product_line,
    "Product_Line",
    "Product_Line",
    "product_line_in_dictionary",
    "error",
    "policies",
    paths_silver
)
```

```python
# Cell 9 — Run metrics & store to _metrics Delta

from datetime import datetime

def write_metric(name: str, value, context: str, paths_silver: dict):
    metrics_path = paths_silver["_metrics"]

    df_metric = spark.createDataFrame(
        [(name, str(value), context, datetime.utcnow())],
        "metric STRING, value STRING, context STRING, ts TIMESTAMP"
    )

    (df_metric.write
        .format("delta")
        .mode("append")
        .save(metrics_path))

    print(f"✅ metric written → {name}={value}, context={context}, path={metrics_path}")

# Write a couple of key metrics
write_metric("rowcount_policies_silver",
             pol_silver.count(),
             "policies_silver",
             paths_silver)

write_metric("distinct_policy_ids_silver",
             pol_silver.select("Policy_ID").distinct().count(),
             "policies_silver",
             paths_silver)

# Read back metrics
metrics_df = spark.read.format("delta").load(paths_silver["_metrics"])
metrics_df.orderBy(F.col("ts").desc()).show(truncate=False)
```

```python
from pyspark.sql import functions as F

# Load Bronze & Silver policies from paths
bronze_policies = spark.read.format("delta").load(paths_bronze["policies"])
silver_policies = spark.read.format("delta").load(paths_silver["policies"])

print("Bronze row count: ", bronze_policies.count())
print("Silver row count: ", silver_policies.count())

metrics_A = {
    "rows_bronze": bronze_policies.count(),
    "rows_silver": silver_policies.count(),
    "distinct_policy_id_bronze": bronze_policies.select("Policy_ID").distinct().count(),
    "distinct_policy_id_silver": silver_policies.select("Policy_ID").distinct().count(),
}

print("A) Snapshot metrics:")
for k, v in metrics_A.items():
    print(f"  {k}: {v}")
```

# 🅱 Cell B — Data quality improvement (basic checks)

```python
def pct(x, n):
    return 0.0 if n == 0 else round(100.0 * x / n, 4)

br = bronze_policies
sv = silver_policies

n_br = br.count()
n_sv = sv.count()

# --- Bronze ---
br_null_keys = br.filter(F.col("Policy_ID").isNull() | F.col("Customer_ID").isNull()).count()
br_bad_dates = br.filter(
    F.col("Policy_Start_Date").isNotNull() &
    F.col("Policy_End_Date").isNotNull() &
    (F.col("Policy_End_Date") < F.col("Policy_Start_Date"))
).count()

br_neg_money = br.filter(
    (F.col("Sum_Insured_GBP") < 0) | (F.col("Annual_Premium_GBP") < 0)
).count()

br_discount_out = br.filter(
    F.col("Discount_Offered_%").isNotNull() &
    ((F.col("Discount_Offered_%") < 0) | (F.col("Discount_Offered_%") > 100))
).count()

# --- Silver ---
sv_null_keys = sv.filter(F.col("Policy_ID").isNull() | F.col("Customer_ID").isNull()).count()
sv_bad_dates = sv.filter(
    F.col("Policy_Start_Date").isNotNull() &
    F.col("Policy_End_Date").isNotNull() &
    (F.col("Policy_End_Date") < F.col("Policy_Start_Date"))
).count()

sv_neg_money = sv.filter(
    (F.col("Sum_Insured_GBP") < 0) | (F.col("Annual_Premium_GBP") < 0)
).count()

sv_discount_out = sv.filter(
    F.col("Discount_Offered_%").isNotNull() &
    ((F.col("Discount_Offered_%") < 0) | (F.col("Discount_Offered_%") > 100))
).count()

dq_report = {
    "bronze_null_keys_pct":        pct(br_null_keys, n_br),
    "silver_null_keys_pct":        pct(sv_null_keys, n_sv),
    "bronze_bad_date_order_pct":   pct(br_bad_dates, n_br),
    "silver_bad_date_order_pct":   pct(sv_bad_dates, n_sv),
    "bronze_negative_money_pct":   pct(br_neg_money, n_br),
    "silver_negative_money_pct":   pct(sv_neg_money, n_sv),
    "bronze_discount_out_pct":     pct(br_discount_out, n_br),
    "silver_discount_out_pct":     pct(sv_discount_out, n_sv),
}

print("B) Data quality improvement (Bronze → Silver):")
for k, v in dq_report.items():
    print(f"  {k}: {v}%")
```

# 🅲 Cell C — Feature differences (incl. new Silver columns + business purpose)

```python
# Column sets
bronze_cols = set(bronze_policies.columns)
silver_cols = set(silver_policies.columns)

new_features      = sorted(list(silver_cols - bronze_cols))   # only in Silver
dropped_features  = sorted(list(bronze_cols - silver_cols))   # only in Bronze
common_features   = sorted(list(bronze_cols & silver_cols))   # shared

print("C1) Column comparison:")
print("  New in Silver:   ", new_features)
print("  Dropped in Silver:", dropped_features)
print("  Common columns:  ", len(common_features))

# Data types
bronze_types = dict(bronze_policies.dtypes)
silver_types = dict(silver_policies.dtypes)

# Align on Policy_ID for accurate per-column comparison
bronze_aligned = bronze_policies.select("Policy_ID", *common_features).dropDuplicates(["Policy_ID"])
silver_aligned = silver_policies.select("Policy_ID", *common_features).dropDuplicates(["Policy_ID"])

joined = (
    bronze_aligned.alias("bz")
    .join(silver_aligned.alias("sv"), on="Policy_ID", how="outer")
)

print("\nC2) Per-column type & value changes for common columns:")
for col in common_features:
    bronze_type = bronze_types.get(col)
    silver_type = silver_types.get(col)
    type_changed = bronze_type != silver_type
    type_msg = (
        f"Type changed: {bronze_type} → {silver_type}"
        if type_changed else
        f"Type unchanged: {bronze_type}"
    )

    n_diff = (
        joined
        .filter(
            (joined[f"bz.{col}"].isNull() & joined[f"sv.{col}"].isNotNull()) |
            (joined[f"bz.{col}"].isNotNull() & joined[f"sv.{col}"].isNull()) |
            (joined[f"bz.{col}"] != joined[f"sv.{col}"])
        )
        .count()
    )

    if n_diff == 0 and not type_changed:
        print(f"  • '{col}' unchanged. {type_msg}")
    else:
        print(f"  • '{col}' transformed in {n_diff} policies. {type_msg}")
        sample_diff = (
            joined
            .filter(
                (joined[f"bz.{col}"].isNull() & joined[f"sv.{col}"].isNotNull()) |
                (joined[f"bz.{col}"].isNotNull() & joined[f"sv.{col}"].isNull()) |
                (joined[f"bz.{col}"] != joined[f"sv.{col}"])
            )
            .select("Policy_ID", F.col(f"bz.{col}").alias("bronze_value"),
                                F.col(f"sv.{col}").alias("silver_value"))
            .limit(5)
        )
        print(f"    Sample differences for '{col}':")
        sample_diff.show(truncate=False)

# ---- NEW: business-purpose explanations for *new* Silver features ----

feature_explanations = {
    # typical Silver enrichments for policies
    "Policy_Duration_Days": (
        "Number of days between Policy_Start_Date and Policy_End_Date. "
        "Used for contract tenure analytics and rate-per-day KPIs."
    ),
    "Renewal_Conversion": (
        "Derived flag where we only keep Renewal_Accepted_Flag when a renewal was actually offered. "
        "Used to calculate clean renewal conversion rates."
    ),
    "dq_money_valid": (
        "Data quality flag: 1 when monetary fields are non-negative. "
        "Helps monitor and explain outliers / bad source records."
    ),
    "dq_discount_valid": (
        "Data quality flag: 1 when Discount_Offered_% is between 0 and 100 or null. "
        "Prevents nonsense discount values from polluting downstream analytics."
    ),
    "dq_renewal_valid": (
        "Data quality flag: 1 when 'accepted' implies 'offered'. "
        "Captures logically inconsistent renewal records."
    ),
    "record_hash": (
        "SHA-256 hash of the row contents. Used for change data capture, de-duplication, "
        "and lineage/audit purposes."
    ),
    "created_at": (
        "Timestamp when this Silver record was created. Supports data freshness reporting and auditing."
    ),
    "source_system": (
        "Identifier for the upstream system or pipeline. Used for lineage, recon, and debugging."
    ),
}

print("\nC3) New Silver-only features (with business purpose):")
for col in new_features:
    reason = feature_explanations.get(
        col,
        "No specific description mapped; this is a Silver-only enrichment or technical column."
    )
    print(f"\n  ➕ '{col}' (business purpose): {reason}")
    # Show sample values
    silver_policies.select("Policy_ID", col).limit(5).show(truncate=False)
    
print("\nC4) Features dropped from Bronze in Silver (usually technical / redundant):")
for col in dropped_features:
    print(f"  ✂ '{col}' was dropped in Silver (likely redundant, low quality, or replaced by a derived field).")
```
