# 🔢 Cell 2 – Read Silver Claims & Providers

```python
# Cell 2 — Read Silver Claims & Providers

SILVER_CLAIMS_PATH    = paths_silver["claims"]
SILVER_PROVIDERS_PATH = paths_silver["providers"]

claims_silver = spark.read.format("delta").load(SILVER_CLAIMS_PATH)
providers_silver = spark.read.format("delta").load(SILVER_PROVIDERS_PATH)

print("Silver Claims schema:")
claims_silver.printSchema()

print("Silver Providers schema:")
providers_silver.printSchema()

print("Claims rowcount:", claims_silver.count())
print("Providers rowcount:", providers_silver.count())

claims_silver.show(5, truncate=False)
providers_silver.show(5, truncate=False)
```

# 🔢 Cell 3 – Define Gold schema & base selection

We’ll build fact_claims as:

one row per Claim_ID

enriched with:

provider-level fraud flag

simple claim KPIs (loss ratio, high cost flag)

DQ flags (money/date)

```python
# Cell 3 — Define target Gold schema & base DF

# Expected schema for Gold fact_claims
fact_claims_schema = StructType([
    StructField("Claim_ID",          StringType(),  True),
    StructField("Provider_ID",       StringType(),  True),
    StructField("Member_Key",        StringType(),  True),
    StructField("Date_Reported",     DateType(),    True),
    StructField("Date_Settled",      DateType(),    True),
    StructField("Payout_GBP",        DoubleType(),  True),
    StructField("Claim_Amount_GBP",  DoubleType(),  True),
    StructField("Fraud_Label",       IntegerType(), True),
    StructField("Claim_Type",        StringType(),  True),
    StructField("Claim_Status",      StringType(),  True),
    # Provider info
    StructField("Provider_Fraud_Flag", IntegerType(), True),
    # Derived metrics
    StructField("Days_To_Settle",     IntegerType(), True),
    StructField("Payout_to_Amount_Ratio", DoubleType(), True),
    StructField("High_Cost_Claim_Flag",  IntegerType(), True),
    # DQ flags carried from Silver
    StructField("dq_money_valid",     IntegerType(), True),
    StructField("dq_date_valid",      IntegerType(), True)
])

# Base select from Silver Claims
claims_base = claims_silver.select(
    F.col("Claim_ID"),
    F.col("Provider_ID"),
    F.col("Member_Key"),
    F.col("Date_Reported"),
    F.col("Date_Settled"),
    F.col("Payout_GBP").cast("double"),
    F.col("Claim_Amount_GBP").cast("double"),
    F.col("Fraud_Label").cast("int"),
    F.col("Claim_Type"),
    F.col("Claim_Status"),
    F.col("Days_To_Settle").cast("int"),
    F.col("dq_money_valid").cast("int"),
    F.col("dq_date_valid").cast("int")
)

claims_base.show(5, truncate=False)
print("Base claims rowcount:", claims_base.count())
```

# 🔢 Cell 4 – Join with Providers & derive metrics

```python
# Cell 4 — Join with Providers & derive business metrics

# Providers: ensure schema
providers_sel = providers_silver.select(
    F.col("Provider_ID").cast("string"),
    F.col("Fraud_Flag").cast("int").alias("Provider_Fraud_Flag")
)

# Left join claims → providers (we keep all claims; provider may be null)
claims_joined = (
    claims_base.alias("c")
    .join(
        providers_sel.alias("p"),
        on="Provider_ID",
        how="left"
    )
)

# Derived metrics
claims_enriched = (
    claims_joined
    # Payout_to_Amount_Ratio: safe division
    .withColumn(
        "Payout_to_Amount_Ratio",
        F.when(
            (F.col("Claim_Amount_GBP").isNotNull()) & (F.col("Claim_Amount_GBP") > 0),
            F.col("Payout_GBP") / F.col("Claim_Amount_GBP")
        ).otherwise(None)
    )
    # High_Cost_Claim_Flag: top ~5–10% claims by payout
    .withColumn(
        "High_Cost_Claim_Flag",
        F.when(F.col("Payout_GBP") >= F.expr("percentile_approx(Payout_GBP, 0.9) OVER ()"), 1).otherwise(0)
    )
)

# Conform to target schema
claims_fact = enforce_schema(claims_enriched, fact_claims_schema)

print("Fact claims schema (after enforce_schema):")
claims_fact.printSchema()
claims_fact.show(10, truncate=False)
print("Fact claims rowcount:", claims_fact.count())
```

🔎 Note: percentile_approx(... ) OVER () works in Spark 3.x.
If you ever hit an issue, we can replace with a precomputed threshold via approxQuantile.

# 🔢 Cell 5 – DQ checks (Gold-level expectations)

We’ll reuse dq_expect from utils_silver:

```python
# Cell 5 — DQ checks on Gold fact_claims

from utils_silver import dq_expect

# 1) Claim_ID must not be null
dq_expect(
    df=claims_fact,
    name="pk_not_null",
    expr="Claim_ID IS NOT NULL",
    severity="error",
    table_label="fact_claims",
    paths_silver=paths_silver
)

# 2) Monetary sanity
dq_expect(
    df=claims_fact,
    name="money_non_negative",
    expr="coalesce(Payout_GBP,0) >= 0 AND coalesce(Claim_Amount_GBP,0) >= 0",
    severity="error",
    table_label="fact_claims",
    paths_silver=paths_silver
)

# 3) Ratio sanity (allow up to 1.5× to be safe)
dq_expect(
    df=claims_fact,
    name="ratio_sane",
    expr="Payout_to_Amount_Ratio IS NULL OR (Payout_to_Amount_Ratio >= 0 AND Payout_to_Amount_Ratio <= 1.5)",
    severity="warn",   # we only warn here, not fail the pipeline
    table_label="fact_claims",
    paths_silver=paths_silver
)

print("DQ checks complete.")
```

# 🔢 Cell 6 – Write Gold fact_claims (Delta + table)

```python
# Cell 6 — Write Gold fact_claims to Delta + register table

print("Writing Gold fact_claims to:", GOLD_CLAIMS_PATH)

(
    claims_fact
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(GOLD_CLAIMS_PATH)
)

print("✅ Gold fact_claims written to Delta path.")

full_table_name = f"{DB_GOLD}.fact_claims"

spark.sql(f"""
CREATE TABLE IF NOT EXISTS {full_table_name}
USING DELTA
LOCATION '{GOLD_CLAIMS_PATH}'
""")

print("✅ Registered Gold table:", full_table_name)

# Quick peek
spark.table(full_table_name).show(10, truncate=False)
```

# 🔢 Cell 9 (A/B/C/D) – Compare Silver vs Gold

Like we did for Policies/Members/Claims Silver, this is your analysis cell.

```python
# Cell 9 — A/B/C/D comparison: Silver Claims vs Gold fact_claims

silver = claims_silver
gold   = spark.table(f"{DB_GOLD}.fact_claims")

# A) Snapshot metrics
print("A) Snapshot:")
print("  Silver rows:", silver.count())
print("  Gold rows:  ", gold.count())

print("  Distinct Claim_ID Silver:", silver.select("Claim_ID").distinct().count())
print("  Distinct Claim_ID Gold:  ", gold.select("Claim_ID").distinct().count())

# B) Check key coverage (are all Silver claims present in Gold?)
silver_only = (
    silver.select("Claim_ID").dropDuplicates()
    .join(gold.select("Claim_ID").dropDuplicates(), on="Claim_ID", how="left_anti")
)

gold_only = (
    gold.select("Claim_ID").dropDuplicates()
    .join(silver.select("Claim_ID").dropDuplicates(), on="Claim_ID", how="left_anti")
)

print("B) Key coverage:")
print("  Claims in Silver but not in Gold:", silver_only.count())
print("  Claims in Gold but not in Silver:", gold_only.count())

# C) Compare core numeric fields (Payout & Amount)
core_cols = ["Payout_GBP", "Claim_Amount_GBP"]

joined = (
    silver.select("Claim_ID", *core_cols).alias("s")
    .join(gold.select("Claim_ID", *core_cols).alias("g"), on="Claim_ID", how="inner")
)

for c in core_cols:
    diff_count = joined.filter(
        (F.col(f"s.{c}").isNull() & F.col(f"g.{c}").isNotNull()) |
        (F.col(f"s.{c}").isNotNull() & F.col(f"g.{c}").isNull()) |
        (F.col(f"s.{c}") != F.col(f"g.{c}"))
    ).count()
    print(f"C) Field '{c}' differences between Silver and Gold:", diff_count)

# D) Sample a few Gold rows to see provider fraud & KPIs
print("D) Sample Gold rows with Provider_Fraud_Flag and KPIs:")
gold.select(
    "Claim_ID",
    "Provider_ID",
    "Payout_GBP",
    "Claim_Amount_GBP",
    "Provider_Fraud_Flag",
    "Payout_to_Amount_Ratio",
    "High_Cost_Claim_Flag",
    "dq_money_valid",
    "dq_date_valid"
).show(10, truncate=False)
```

# 🔢 Cell 10 – Simple Gold KPIs (for dashboards / storytelling)

```python
# Cell 10 — Simple Gold KPIs for dashboards or interviews

gold = spark.table(f"{DB_GOLD}.fact_claims")

# 1) Average payout and amount by claim type
gold.groupBy("Claim_Type").agg(
    F.count("*").alias("n_claims"),
    F.avg("Payout_GBP").alias("avg_payout"),
    F.avg("Claim_Amount_GBP").alias("avg_claim_amount")
).orderBy(F.desc("n_claims")).show(truncate=False)

# 2) Fraud vs non-fraud (using claim-level Fraud_Label)
gold.groupBy("Fraud_Label").agg(
    F.count("*").alias("n_claims"),
    F.avg("Payout_GBP").alias("avg_payout"),
    F.avg("Claim_Amount_GBP").alias("avg_claim_amount")
).show(truncate=False)

# 3) High-cost claim ratio by status
gold.groupBy("Claim_Status").agg(
    F.count("*").alias("n_claims"),
    F.sum("High_Cost_Claim_Flag").alias("n_high_cost_claims")
).withColumn(
    "high_cost_ratio",
    F.col("n_high_cost_claims") / F.col("n_claims")
).show(truncate=False)

print("✅ Gold fact_claims ready for BI / ML.")
```
