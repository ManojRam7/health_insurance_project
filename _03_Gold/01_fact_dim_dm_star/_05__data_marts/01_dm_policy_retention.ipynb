# 📓 Cell 1 – Imports & config

```python
from pyspark.sql import functions as F

DB_GOLD = "bupa_gold"
storage_account = "clientdatastorage"
GOLD_CONTAINER = "golddata"

GOLD_BASE = f"abfss://{GOLD_CONTAINER}@{storage_account}.dfs.core.windows.net"
FACT_POLICIES_PATH = f"{GOLD_BASE}/fact_policies"
DM_POLICY_RETENTION_PATH = f"{GOLD_BASE}/dm_policy_retention"

print("GOLD_BASE:", GOLD_BASE)
print("FACT_POLICIES_PATH:", FACT_POLICIES_PATH)
print("DM_POLICY_RETENTION_PATH:", DM_POLICY_RETENTION_PATH)
```

# 📓 Cell 2 – Read base fact table

We’ll trust the already-built fact_policies as our grain.

```python
fact_policies = spark.read.format("delta").load(FACT_POLICIES_PATH)

print("fact_policies rows:", fact_policies.count())
fact_policies.printSchema()
fact_policies.show(5, truncate=False)
```

You should see columns like:

Policy_ID

Customer_ID

Product_Line

Channel

Sum_Insured_GBP

Annual_Premium_GBP

Policy_Start_Date, Policy_End_Date, Policy_Duration_Days

Renewal_Offered_Flag, Renewal_Accepted_Flag, Renewal_Conversion

Tenure_Band, Premium_Band, Discount_Band

dq_money_valid, dq_discount_valid, dq_renewal_valid

# 📓 Cell 3 – Add date attributes & filter to “trusted” policies

For a business mart we usually restrict to data that passes DQ checks.

```python
# 1) Date attributes
fact_base = (
    fact_policies
    .withColumn("Policy_Start_Year",  F.year("Policy_Start_Date"))
    .withColumn("Policy_Start_Month", F.date_format("Policy_Start_Date", "yyyy-MM"))
)

# 2) Trusted slice: all core DQ flags = 1
fact_trusted = (
    fact_base
    .filter(
        (F.col("dq_money_valid") == 1) &
        (F.col("dq_discount_valid") == 1) &
        (F.col("dq_renewal_valid") == 1)
    )
)

print("Total fact_policies rows      :", fact_policies.count())
print("Trusted policies (all dq=1)   :", fact_trusted.count())
fact_trusted.select(
    "Policy_ID","Product_Line","Channel",
    "Policy_Start_Date","Policy_Duration_Days",
    "Annual_Premium_GBP","Renewal_Offered_Flag","Renewal_Accepted_Flag",
    "dq_money_valid","dq_discount_valid","dq_renewal_valid"
).show(10, truncate=False)
```

# 📓 Cell 4 – Build the Policy Retention mart (aggregations)

Grain:
Product_Line, Channel, Tenure_Band, Policy_Start_Year

Metrics:

Policy_Count

Total_Premium_GBP, Avg_Premium_GBP

Offer_Rate (how many got a renewal offer)

Renewal_Rate (how many accepted when offered)

Clean_Renewal_Conversion (Renewal_Conversion using trusted records)

```python
agg_exprs = {
    "Policy_ID": "count",
    "Annual_Premium_GBP": "sum"
}

fact_group = (
    fact_trusted
    .groupBy(
        "Product_Line",
        "Channel",
        "Tenure_Band",
        "Policy_Start_Year"
    )
    .agg(
        F.count("Policy_ID").alias("Policy_Count"),
        F.sum("Annual_Premium_GBP").alias("Total_Premium_GBP"),
        F.avg("Annual_Premium_GBP").alias("Avg_Premium_GBP"),
        F.avg(F.col("Renewal_Offered_Flag").cast("double")).alias("Offer_Rate"),
        F.avg(F.col("Renewal_Accepted_Flag").cast("double")).alias("Acceptance_Rate"),
        F.avg(F.col("Renewal_Conversion").cast("double")).alias("Clean_Renewal_Conversion")
    )
)

dm_policy_retention = (
    fact_group
    .withColumn("Offer_Rate",               F.round("Offer_Rate", 4))
    .withColumn("Acceptance_Rate",          F.round("Acceptance_Rate", 4))
    .withColumn("Clean_Renewal_Conversion", F.round("Clean_Renewal_Conversion", 4))
    .withColumn("Total_Premium_GBP",        F.round("Total_Premium_GBP", 2))
    .withColumn("Avg_Premium_GBP",          F.round("Avg_Premium_GBP", 2))
)

dm_policy_retention.orderBy(
    "Product_Line","Channel","Tenure_Band","Policy_Start_Year"
).show(20, truncate=False)
dm_policy_retention.printSchema()
```

# 📓 Cell 5 – Write mart to golddata and register in bupa_gold

```python
# Create the Gold schema if it doesn't exist
spark.sql("CREATE DATABASE IF NOT EXISTS bupa_gold")

print("Available databases:")
spark.sql("SHOW DATABASES").show()
```

```python
# 1) Write to Gold container
(dm_policy_retention.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(DM_POLICY_RETENTION_PATH))

print("✅ dm_policy_retention written to:", DM_POLICY_RETENTION_PATH)

# 2) Register as a Gold table
spark.sql(f"DROP TABLE IF EXISTS {DB_GOLD}.dm_policy_retention")

spark.sql(f"""
CREATE TABLE {DB_GOLD}.dm_policy_retention
USING DELTA
LOCATION '{DM_POLICY_RETENTION_PATH}'
""")

print(f"✅ Registered table: {DB_GOLD}.dm_policy_retention")

# 3) Sanity check from SQL
spark.table(f"{DB_GOLD}.dm_policy_retention").show(10, truncate=False)
```

# 📓 Cell 6 (A) – Quick analysis: Top segments by renewal conversion

```python
dm = spark.table(f"{DB_GOLD}.dm_policy_retention")

dm.orderBy(F.col("Clean_Renewal_Conversion").desc()).show(20, truncate=False)
```

# 📓 Cell 7 (B) – Analysis: Premium vs retention by product

```python
dm.groupBy("Product_Line").agg(
    F.sum("Policy_Count").alias("Policies"),
    F.round(F.avg("Clean_Renewal_Conversion"), 4).alias("Avg_Renewal_Conv"),
    F.round(F.avg("Avg_Premium_GBP"), 2).alias("Avg_Premium")
).orderBy("Product_Line").show(truncate=False)
```

# 📓 Cell 8 (C) – Analysis: Channel performance

```python
dm.groupBy("Channel").agg(
    F.sum("Policy_Count").alias("Policies"),
    F.round(F.avg("Clean_Renewal_Conversion"), 4).alias("Avg_Renewal_Conv"),
    F.round(F.avg("Offer_Rate"), 4).alias("Avg_Offer_Rate")
).orderBy(F.col("Policies").desc()).show(truncate=False)
```

# 📓 Cell 9 (D) – Analysis: Tenure band insight

```python
dm.groupBy("Tenure_Band").agg(
    F.sum("Policy_Count").alias("Policies"),
    F.round(F.avg("Clean_Renewal_Conversion"), 4).alias("Avg_Renewal_Conv"),
    F.round(F.avg("Avg_Premium_GBP"), 2).alias("Avg_Premium")
).orderBy("Tenure_Band").show(truncate=False)
```

# 📓 Cell 10 – Simple “gold KPIs” view for dashboards

This gives you a shortcut view you can point Power BI / Tableau at.

```python
kpi_view = (
    dm_policy_retention
    .groupBy("Product_Line","Channel")
    .agg(
        F.sum("Policy_Count").alias("Policies"),
        F.round(F.avg("Clean_Renewal_Conversion"), 4).alias("Renewal_Conv"),
        F.round(F.avg("Avg_Premium_GBP"), 2).alias("Avg_Premium")
    )
)

kpi_view.orderBy(F.col("Policies").desc()).show(20, truncate=False)
```

```python
from pyspark.sql import functions as F

DB_GOLD = "bupa_gold"
KPI_TABLE = f"{DB_GOLD}.vw_policy_kpi"   # it's a table, name still says vw_ if you like

# Just write the KPI DataFrame as a Delta table
(
    kpi_view
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(KPI_TABLE)
)

print(f"✅ Registered KPI table: {KPI_TABLE}")

# Quick sanity check
spark.table(KPI_TABLE).show(10, truncate=False)
```
