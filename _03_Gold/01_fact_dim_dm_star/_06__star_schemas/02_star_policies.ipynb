# Cell 1 — Setup & load base tables

```python
# Cell 1 — Setup & load base tables for star_policies

from pyspark.sql import functions as F

DB_GOLD        = "bupa_gold"
GOLD_ACCOUNT   = "clientdatastorage"
GOLD_CONTAINER = "golddata"

STAR_POLICIES_PATH = (
    f"abfss://{GOLD_CONTAINER}@{GOLD_ACCOUNT}.dfs.core.windows.net/star_policies"
)

# Make sure DB exists (no-op if already created)
spark.sql(f"CREATE DATABASE IF NOT EXISTS {DB_GOLD}")

# Load fact + dimensions from Gold
fact_policies = spark.table(f"{DB_GOLD}.fact_policies")
dim_channel   = spark.table(f"{DB_GOLD}.dim_channel")
dim_product   = spark.table(f"{DB_GOLD}.dim_product_line")

print("fact_policies rowcount:", fact_policies.count())
fact_policies.limit(5).show(truncate=False)

print("dim_channel:")
dim_channel.limit(5).show(truncate=False)

print("dim_product_line:")
dim_product.limit(5).show(truncate=False)
```

# Cell 2 — Build star_policies DataFrame (join fact + dims)

```python
# Cell 2 — Build star_policies (one row per policy, enriched with dims)

fp = fact_policies.alias("fp")
dc = dim_channel.alias("dc")
dp = dim_product.alias("dp")

# Join on natural business keys
star_base = (
    fp
    .join(dp, fp.Product_Line == dp.Product_Line, how="left")
    .join(dc, fp.Channel     == dc.Channel_Name, how="left")
)

print("star_base rowcount:", star_base.count())

# Select & rename columns for star schema
star_policies = star_base.select(
    # Business grain
    F.col("fp.Policy_ID").alias("Policy_ID"),
    F.col("fp.Customer_ID").alias("Customer_ID"),

    # Product dimension
    F.col("dp.Product_Line_Code").alias("Product_Line_Code"),
    F.col("dp.Product_Line").alias("Product_Line_Name"),

    # Channel dimension
    F.col("dc.Channel_Code").alias("Channel_Code"),
    F.col("dc.Channel_Name").alias("Channel_Name"),

    # Core policy measures
    F.col("fp.Sum_Insured_GBP"),
    F.col("fp.Annual_Premium_GBP"),
    F.col("fp.Policy_Start_Date"),
    F.col("fp.Policy_End_Date"),
    F.col("fp.Policy_Duration_Days"),

    # Renewal behaviour
    F.col("fp.Renewal_Offered_Flag"),
    F.col("fp.Renewal_Accepted_Flag"),
    F.col("fp.Renewal_Conversion"),

    # Feature / KPI bands
    F.col("fp.Tenure_Band"),
    F.col("fp.Premium_Band"),
    F.col("fp.Discount_Band"),
    F.col("fp.Renewal_Outcome"),

    # DQ flags
    F.col("fp.dq_money_valid"),
    F.col("fp.dq_discount_valid"),
    F.col("fp.dq_renewal_valid")
)

star_policies.printSchema()
star_policies.show(10, truncate=False)
```

# Cell 3 — Write Delta table + register in bupa_gold

```python
# Cell 3 — Write star_policies to golddata + register Gold table

print("Writing star_policies to:", STAR_POLICIES_PATH)

(star_policies
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(STAR_POLICIES_PATH)
)

print("✅ Wrote Delta files for star_policies.")

# Register as Gold table
spark.sql("DROP TABLE IF EXISTS bupa_gold.star_policies")

spark.sql(f"""
CREATE TABLE bupa_gold.star_policies
USING DELTA
LOCATION '{STAR_POLICIES_PATH}'
""")

print("✅ Registered table: bupa_gold.star_policies")
```

# Cell A — Snapshot & schema

```python
# Cell A — Snapshot & schema for star_policies

star_p = spark.table("bupa_gold.star_policies")

print(f"star_policies rowcount: {star_p.count():,}")
star_p.printSchema()

star_p.select(
    "Policy_ID",
    "Customer_ID",
    "Product_Line_Code",
    "Product_Line_Name",
    "Channel_Code",
    "Channel_Name",
    "Annual_Premium_GBP",
    "Policy_Duration_Days",
    "Renewal_Outcome"
).show(10, truncate=False)
```

# Cell B — Product × Channel profitability view

```python
# Cell B — Performance by Product_Line and Channel

from pyspark.sql import functions as F

prod_channel_summary = (
    star_p.groupBy("Product_Line_Name", "Channel_Name")
          .agg(
              F.count("*").alias("Policy_Count"),
              F.sum("Annual_Premium_GBP").alias("Total_Premium_GBP"),
              F.sum("Sum_Insured_GBP").alias("Total_Sum_Insured_GBP"),
              F.avg("Annual_Premium_GBP").alias("Avg_Premium_GBP")
          )
          .orderBy(F.col("Policy_Count").desc())
)

prod_channel_summary.show(30, truncate=False)
```

# Cell C — Renewal & discount behaviour

```python
# Cell C — Renewal rate & discount patterns

renewal_stats = (
    star_p.groupBy("Product_Line_Name", "Channel_Name", "Renewal_Outcome")
          .agg(
              F.count("*").alias("Policy_Count"),
              F.avg("Discount_Band").alias("Discount_Band_avg").alias("Discount_Band_avg")  # if Discount_Band is string, we won't avg
          )
)

renewal_stats.show(50, truncate=False)

# Simpler KPI: overall renewal rate by Product_Line
renewal_rate_pl = (
    star_p.groupBy("Product_Line_Name")
          .agg(
              F.count("*").alias("Policies"),
              F.sum(F.when(F.col("Renewal_Outcome") == "Renewed", 1).otherwise(0)).alias("Renewed"),
          )
          .withColumn("Renewal_Rate_pct", (F.col("Renewed") / F.col("Policies") * 100.0))
          .orderBy(F.col("Renewal_Rate_pct").desc())
)

renewal_rate_pl.show(truncate=False)
```

# Cell D — Data quality lens on star_policies

```python
# Cell D — DQ coverage for star_policies

total = star_p.count()

dq_summary = (
    star_p.select(
        F.sum(F.when(F.col("dq_money_valid") == 1, 1).otherwise(0)).alias("money_valid"),
        F.sum(F.when(F.col("dq_discount_valid") == 1, 1).otherwise(0)).alias("discount_valid"),
        F.sum(F.when(F.col("dq_renewal_valid") == 1, 1).otherwise(0)).alias("renewal_valid")
    )
    .withColumn("total_rows", F.lit(total))
)

dq_summary.show(truncate=False)
```
