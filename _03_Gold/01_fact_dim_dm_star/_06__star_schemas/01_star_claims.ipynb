# Cell 1 – Imports & config

```python
from pyspark.sql import functions as F

DB_GOLD = "bupa_gold"
STORAGE_ACCOUNT = "clientdatastorage"
GOLD_CONTAINER = "golddata"

STAR_CLAIMS_PATH = f"abfss://{GOLD_CONTAINER}@{STORAGE_ACCOUNT}.dfs.core.windows.net/star_claims"

print("STAR_CLAIMS_PATH:", STAR_CLAIMS_PATH)
```

# Cell 2 – Load fact + dimension tables

```python
fact_claims = spark.table(f"{DB_GOLD}.fact_claims")
dim_claim_type = spark.table(f"{DB_GOLD}.dim_claim_type")
dim_providers  = spark.table(f"{DB_GOLD}.dim_providers")

print("fact_claims rowcount:", fact_claims.count())
fact_claims.show(5, truncate=False)

print("dim_claim_type:")
dim_claim_type.show(5, truncate=False)

print("dim_providers:")
dim_providers.show(5, truncate=False)
```

# Cell 3 – Build star_claims DataFrame

join dim_claim_type on the text Claim_Type

join dim_providers on Provider_ID

avoid ambiguous columns by using aliases

```python
fc = fact_claims.alias("fc")
dct = dim_claim_type.alias("dct")
dp  = dim_providers.alias("dp")

star_claims_df = (
    fc
    # join claim type dimension (on natural name)
    .join(
        dct,
        on=F.upper(F.col("fc.Claim_Type")) == F.upper(F.col("dct.Claim_Type")),
        how="left"
    )
    # join provider dimension (on Provider_ID)
    .join(
        dp,
        on=F.col("fc.Provider_ID") == F.col("dp.Provider_ID"),
        how="left"
    )
    .select(
        # Degenerate / business keys
        F.col("fc.Claim_ID").alias("Claim_ID"),
        F.col("fc.Member_Key").alias("Member_Key"),
        F.col("fc.Provider_ID").alias("Provider_ID"),

        # Dates & measures
        F.col("fc.Date_Reported"),
        F.col("fc.Date_Settled"),
        F.col("fc.Payout_GBP"),
        F.col("fc.Claim_Amount_GBP"),
        F.col("fc.Days_To_Settle"),
        F.col("fc.Payout_to_Amount_Ratio"),
        F.col("fc.High_Cost_Claim_Flag"),

        # Claim type & status (from dim + fact)
        F.col("dct.Claim_Type_Code").alias("Claim_Type_Code"),
        F.col("dct.Claim_Type").alias("Claim_Type_Name"),
        F.col("fc.Claim_Type").alias("Claim_Type_Fact"),
        F.col("fc.Claim_Status"),

        # Fraud & risk
        F.col("fc.Fraud_Label").alias("Claim_Fraud_Label"),
        F.col("fc.Provider_Fraud_Flag").alias("Provider_Fraud_Flag_Fact"),
        F.col("dp.Fraud_Flag").alias("Provider_Fraud_Flag_Dim"),
        F.col("dp.Provider_Risk_Tier"),

        # DQ flags
        F.col("fc.dq_money_valid"),
        F.col("fc.dq_date_valid")
    )
)

print("star_claims_df rowcount:", star_claims_df.count())
star_claims_df.show(10, truncate=False)
star_claims_df.printSchema()
```

Notes:

You now have both dim-based attributes (Claim_Type_Code, Provider_Risk_Tier) and fact-based ones.

You can later simplify columns if you want (e.g. drop Claim_Type_Fact and just keep the dimension version).

# Cell 4 – Persist star table in golddata and register table/view

```python
# 1) Write to Delta in GOLD container
(star_claims_df.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(STAR_CLAIMS_PATH))

print("✅ Wrote star_claims to:", STAR_CLAIMS_PATH)

# 2) Register as table in bupa_gold
spark.sql("CREATE DATABASE IF NOT EXISTS bupa_gold")

spark.sql("DROP TABLE IF EXISTS bupa_gold.star_claims")

spark.sql(f"""
CREATE TABLE bupa_gold.star_claims
USING DELTA
LOCATION '{STAR_CLAIMS_PATH}'
""")

print("✅ Registered table: bupa_gold.star_claims")

# 3) (Optional) create a simple BI-friendly view with fewer columns
spark.sql("DROP VIEW IF EXISTS bupa_gold.vw_star_claims")

spark.sql("""
CREATE VIEW bupa_gold.vw_star_claims AS
SELECT
  Claim_ID,
  Member_Key,
  Provider_ID,
  Claim_Type_Code,
  Claim_Type_Name,
  Claim_Status,
  Payout_GBP,
  Claim_Amount_GBP,
  High_Cost_Claim_Flag,
  Claim_Fraud_Label,
  Provider_Risk_Tier
FROM bupa_gold.star_claims
""")

print("✅ Created view: bupa_gold.vw_star_claims")

spark.table("bupa_gold.star_claims").show(5, truncate=False)
spark.table("bupa_gold.vw_star_claims").show(5, truncate=False)
```

# Cell A – Snapshot & schema

```python
# Cell A — Snapshot & schema

star = spark.table("bupa_gold.star_claims")

print(f"star_claims rowcount: {star.count():,}")
star.printSchema()

star.select(
    "Claim_ID",
    "Member_Key",
    "Provider_ID",
    "Claim_Type_Code",
    "Claim_Type_Name",
    "Provider_Risk_Tier",
    "Claim_Amount_GBP",
    "Payout_GBP",
    "High_Cost_Claim_Flag",
    "dq_money_valid",
    "dq_date_valid"
).show(10, truncate=False)
```

# Cell B – Basic profile (by claim type & status)

```python
# Cell B — Profile by Claim_Type and Claim_Status

from pyspark.sql import functions as F

profile_type_status = (
    star.groupBy("Claim_Type_Name", "Claim_Status")
        .agg(
            F.count("*").alias("Claim_Count"),
            F.sum("Claim_Amount_GBP").alias("Total_Claim_Amount_GBP"),
            F.sum("Payout_GBP").alias("Total_Payout_GBP"),
            F.avg("Payout_to_Amount_Ratio").alias("Avg_Payout_Ratio")
        )
        .orderBy(F.col("Claim_Count").desc())
)

profile_type_status.show(20, truncate=False)
```

# Cell C – Provider risk & fraud view

```python
# Cell C — Provider risk & fraud patterns

provider_risk_summary = (
    star.groupBy("Provider_Risk_Tier")
        .agg(
            F.count("*").alias("Claim_Count"),
            F.sum("Payout_GBP").alias("Total_Payout_GBP"),
            F.avg("Payout_to_Amount_Ratio").alias("Avg_Payout_Ratio"),
            F.sum(F.col("Claim_Fraud_Label").cast("int")).alias("Fraud_Claims")
        )
        .orderBy(F.col("Claim_Count").desc())
)

provider_risk_summary.show(truncate=False)
```

# Cell D – Data quality lens on star_claims

```python
# Cell D — DQ coverage in the star

total = star.count()

dq_summary = (
    star.select(
        F.sum(F.when(F.col("dq_money_valid") == 1, 1).otherwise(0)).alias("money_valid"),
        F.sum(F.when(F.col("dq_date_valid") == 1, 1).otherwise(0)).alias("date_valid"),
        F.sum(F.when(F.col("High_Cost_Claim_Flag") == 1, 1).otherwise(0)).alias("high_cost_claims")
    )
    .withColumn("total_rows", F.lit(total))
)

dq_summary.show(truncate=False)
```
