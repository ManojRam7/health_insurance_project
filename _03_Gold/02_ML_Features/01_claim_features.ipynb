# Cell 1 — Imports & Config

```python
# ML Feature Tables (Gold) — Policies & Claims

from pyspark.sql import functions as F

DB_GOLD = "bupa_gold"

# Azure paths for Gold feature tables
STORAGE_ACCOUNT = "clientdatastorage"
GOLD_CONTAINER  = "golddata"

FT_POLICY_CHURN_PATH = f"abfss://{GOLD_CONTAINER}@{STORAGE_ACCOUNT}.dfs.core.windows.net/ft_policy_churn"
FT_CLAIMS_RISK_PATH  = f"abfss://{GOLD_CONTAINER}@{STORAGE_ACCOUNT}.dfs.core.windows.net/ft_claims_risk"

print("DB_GOLD:", DB_GOLD)
print("FT_POLICY_CHURN_PATH:", FT_POLICY_CHURN_PATH)
print("FT_CLAIMS_RISK_PATH :", FT_CLAIMS_RISK_PATH)
```

# Cell 2 — Load Gold Facts & Dimensions

```python
# Fact tables from Gold
fact_policies = spark.table(f"{DB_GOLD}.fact_policies")
fact_claims   = spark.table(f"{DB_GOLD}.fact_claims")

print("fact_policies:", fact_policies.count())
print("fact_claims  :", fact_claims.count())

fact_policies.printSchema()
fact_claims.printSchema()
```

# Cell 3 — Build Policy-Level Churn Feature Base

```python
# Base set of policy features from fact_policies

p = fact_policies.alias("p")

policy_feat_base = (
    p.select(
        "Policy_ID",
        "Customer_ID",
        "Product_Line",
        "Channel",
        "Sum_Insured_GBP",
        "Annual_Premium_GBP",
        "Policy_Start_Date",
        "Policy_End_Date",
        "Policy_Duration_Days",
        "Renewal_Offered_Flag",
        "Renewal_Accepted_Flag",
        "Renewal_Conversion",
        "Tenure_Band",
        "Premium_Band",
        "Discount_Band",
        "Renewal_Outcome",
        "dq_money_valid",
        "dq_discount_valid",
        "dq_renewal_valid"
    )
)

policy_feat_base.show(5, truncate=False)
policy_feat_base.printSchema()
```

# Cell 4 — Add Churn Label & Extra Features

```python
# Define churn label & additional engineered features

policy_features = (
    policy_feat_base
    # Binary: was renewal offered?
    .withColumn(
        "Is_Renewal_Offered",
        F.when(F.col("Renewal_Offered_Flag") == 1, 1).otherwise(0)
    )
    # Binary churn label:
    #   1 = offered but not accepted
    #   0 = offered & accepted
    #   null = no offer / unknown
    .withColumn(
        "Churn_Label",
        F.when(
            (F.col("Renewal_Offered_Flag") == 1) &
            (F.col("Renewal_Accepted_Flag") == 0),
            1
        ).when(
            (F.col("Renewal_Offered_Flag") == 1) &
            (F.col("Renewal_Accepted_Flag") == 1),
            0
        ).otherwise(F.lit(None).cast("int"))
    )
    # Was any discount given? (based on Discount_Band)
    .withColumn(
        "Is_Discounted",
        F.when(F.col("Discount_Band").isin("0–10%", "10–20%", "20%+"), 1).otherwise(0)
    )
    # Simple numeric feature: premium per £1k of sum insured
    .withColumn(
        "Premium_per_1k_SumInsured",
        F.when(F.col("Sum_Insured_GBP") > 0,
               F.col("Annual_Premium_GBP") / (F.col("Sum_Insured_GBP") / 1000.0)
        ).otherwise(None)
    )
)

policy_features.show(10, truncate=False)
policy_features.groupBy("Churn_Label").count().show()
```

# Cell 5 — Write Policy Churn Feature Table to Gold

```python
# Write as Delta to golddata, then register as ft_policy_churn

print("Writing ft_policy_churn to:", FT_POLICY_CHURN_PATH)

(policy_features
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(FT_POLICY_CHURN_PATH)
)

print("✅ Delta files written.")

spark.sql(f"DROP TABLE IF EXISTS {DB_GOLD}.ft_policy_churn")

spark.sql(f"""
CREATE TABLE {DB_GOLD}.ft_policy_churn
USING DELTA
LOCATION '{FT_POLICY_CHURN_PATH}'
""")

print("✅ Registered table:", f"{DB_GOLD}.ft_policy_churn")

spark.table(f"{DB_GOLD}.ft_policy_churn").show(10, truncate=False)
```

# Cell 6 — Build Claim-Level Risk/Fraud Feature Base

```python
# Load dimensions used for claims features
dim_claim_type = spark.table(f"{DB_GOLD}.dim_claim_type")   # Claim_Type_Code, Claim_Type
dim_providers  = spark.table(f"{DB_GOLD}.dim_providers")    # Provider_ID, Fraud_Flag, Provider_Risk_Tier, ...

print("dim_claim_type:", dim_claim_type.count())
print("dim_providers :", dim_providers.count())

dim_claim_type.show(5, truncate=False)
dim_providers.show(5, truncate=False)
```

# Cell 7 — Join Facts + Dimensions & Engineer Claim Features

```python
fc  = fact_claims.alias("fc")
dct = dim_claim_type.alias("dct")
dp  = dim_providers.alias("dp")

claims_joined = (
    fc
    # join on natural key Claim_Type text
    .join(dct, on=fc["Claim_Type"] == dct["Claim_Type"], how="left")
    # join provider risk info
    .join(dp, on=fc["Provider_ID"] == dp["Provider_ID"], how="left")
)

print("claims_joined rowcount:", claims_joined.count())

claims_features = (
    claims_joined.select(
        F.col("fc.Claim_ID").alias("Claim_ID"),
        F.col("fc.Member_Key").alias("Member_Key"),
        F.col("fc.Provider_ID").alias("Provider_ID"),
        F.col("fc.Claim_Type").alias("Claim_Type_Name"),
        F.col("dct.Claim_Type_Code").alias("Claim_Type_Code"),
        F.col("fc.Claim_Status").alias("Claim_Status"),
        F.col("fc.Claim_Amount_GBP"),
        F.col("fc.Payout_GBP"),
        F.col("fc.Payout_to_Amount_Ratio"),
        F.col("fc.Days_To_Settle"),
        F.col("fc.High_Cost_Claim_Flag"),
        F.col("fc.Fraud_Label").alias("Claim_Fraud_Label"),
        F.col("dp.Fraud_Flag").alias("Provider_Fraud_Label"),
        F.col("dp.Provider_Risk_Tier"),
        F.col("fc.dq_money_valid"),
        F.col("fc.dq_date_valid")
    )
    # Overall fraud label: claim-level first, then provider-level flag as backup
    .withColumn(
        "Is_Fraudulent_Claim",
        F.when(F.col("Claim_Fraud_Label") == 1, 1)
         .when(F.col("Provider_Fraud_Label") == 1, 1)
         .otherwise(0)
    )
    # Binary high-cost indicator (ensure int type)
    .withColumn(
        "Is_High_Cost",
        F.col("High_Cost_Claim_Flag").cast("int")
    )
)

claims_features.show(10, truncate=False)
claims_features.groupBy("Is_Fraudulent_Claim").count().show()
```

# Cell 8 — Write Claims Risk Feature Table to Gold

```python
print("Writing ft_claims_risk to:", FT_CLAIMS_RISK_PATH)

(claims_features
    .write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .save(FT_CLAIMS_RISK_PATH)
)

print("✅ Delta files written.")

spark.sql(f"DROP TABLE IF EXISTS {DB_GOLD}.ft_claims_risk")

spark.sql(f"""
CREATE TABLE {DB_GOLD}.ft_claims_risk
USING DELTA
LOCATION '{FT_CLAIMS_RISK_PATH}'
""")

print("✅ Registered table:", f"{DB_GOLD}.ft_claims_risk")

spark.table(f"{DB_GOLD}.ft_claims_risk").show(10, truncate=False)
```

# Cell A — Policy Churn: Basic Profiling

```python
ft_policy = spark.table(f"{DB_GOLD}.ft_policy_churn")

print("Rows:", ft_policy.count())
print("Distinct Policy_ID:", ft_policy.select("Policy_ID").distinct().count())

print("\nChurn rate by Product_Line:")
(
    ft_policy
    .groupBy("Product_Line", "Churn_Label")
    .count()
    .orderBy("Product_Line", "Churn_Label")
).show(100, truncate=False)

print("\nChurn vs Discounted:")
(
    ft_policy
    .groupBy("Is_Discounted", "Churn_Label")
    .count()
    .orderBy("Is_Discounted", "Churn_Label")
).show(truncate=False)
```

# Cell C — Feature Completeness (Null Checks)

```python
def null_pct(df):
    total = df.count()
    return df.select([
        (F.count(F.when(F.col(c).isNull(), c)) / F.lit(max(total, 1))).alias(c)
        for c in df.columns
    ])

print("Null percentages — ft_policy_churn:")
null_pct(ft_policy).show(truncate=False)

print("\nNull percentages — ft_claims_risk:")
null_pct(ft_claims).show(truncate=False)
```

```python
from pyspark.sql import Row

policy_features_desc = [
    Row(feature="Churn_Label",              type="int",    role="label",   note="1 = churned (offer not accepted), 0 = renewed"),
    Row(feature="Product_Line",             type="string", role="feature", note="Insurance product category"),
    Row(feature="Channel",                  type="string", role="feature", note="Sales channel"),
    Row(feature="Sum_Insured_GBP",          type="double", role="feature", note="Exposure / coverage amount"),
    Row(feature="Annual_Premium_GBP",       type="double", role="feature", note="Pricing signal"),
    Row(feature="Policy_Duration_Days",     type="int",    role="feature", note="Contract tenure"),
    Row(feature="Tenure_Band",              type="string", role="feature", note="Binned tenure"),
    Row(feature="Premium_Band",             type="string", role="feature", note="Binned premium"),
    Row(feature="Discount_Band",            type="string", role="feature", note="Binned discount"),
    Row(feature="Is_Discounted",            type="int",    role="feature", note="Any discount applied"),
    Row(feature="Premium_per_1k_SumInsured",type="double", role="feature", note="Relative price"),
]

claims_features_desc = [
    Row(feature="Is_Fraudulent_Claim",      type="int",    role="label",   note="1 = fraud (claim or provider flagged)"),
    Row(feature="Claim_Type_Code",          type="string", role="feature", note="Normalised claim type"),
    Row(feature="Claim_Status",             type="string", role="feature", note="Settled / Pending / Rejected"),
    Row(feature="Claim_Amount_GBP",         type="double", role="feature", note="Original claim amount"),
    Row(feature="Payout_GBP",               type="double", role="feature", note="Amount paid"),
    Row(feature="Payout_to_Amount_Ratio",   type="double", role="feature", note="Severity proxy"),
    Row(feature="Days_To_Settle",           type="int",    role="feature", note="Cycle time"),
    Row(feature="Is_High_Cost",             type="int",    role="feature", note="High-cost flag"),
    Row(feature="Provider_Risk_Tier",       type="string", role="feature", note="Provider risk profile"),
]

print("Policy churn feature dictionary:")
spark.createDataFrame(policy_features_desc).show(truncate=False)

print("\nClaims risk feature dictionary:")
spark.createDataFrame(claims_features_desc).show(truncate=False)
```
